```python
from sklearn.datasets import load_iris
iris=load_iris()
attr=iris.data
print(iris.feature_names)
```

```python
plt.scatter(attr[:,0],attr[:,1],s=80,c='g',marker='^')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
#attr[:,3] ,petal width using another column to size various data points in scatter plot
plt.scatter(attr[:,0],attr[:,1],s= 100*attr[:,3]**2,alpha=0.5)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
#attr[:,3] ,petal width using another column to size various data points in scatter plot
#assigning color as well using target variable values
# we can change line in code block using backslash
target=iris.target
plt.scatter(attr[:,0],attr[:,1],s= 100*attr[:,3]**2,\
            c=target,alpha=0.5)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
# add color bar to indicate values plotted on this plot
#ticks is the position of label on axis or label and .25,.50 etc are the namesof ticks
# we want to only illustarte values we are getting from target so we will put 0,1,2 in ticks
target=iris.target
plt.scatter(attr[:,0],attr[:,1],s= 100*attr[:,3]**2,\
            c=target,alpha=0.5)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
cbar= plt.colorbar(ticks=[0,1,2])
cbar.set_ticklabels(['Setosa','Vericolor','Virginica'])
plt.show()
```

```python
#target=iris.target
select_index=target==0 # generate logical values on whether setosa
setosa=attr[select_index,:] # default number of bins is 10
plt.hist(setosa[:,1])
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.title('Setosa')
plt.show()
```

```python
# plotting two plot for setosa and versicolor here in one plot
versicolor=attr[target==1,:]
num_bins=15
plt.hist(setosa[:,1],num_bins)
plt.hist(versicolor[:,1],num_bins)
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.show()
```

```python
# using alpha to distiguish between two since they were overlapping
versicolor=attr[target==1,:]
num_bins=15
plt.hist(setosa[:,1],num_bins)
plt.hist(versicolor[:,1],num_bins,alpha=0.75)
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.show()
```

```python
# assigning diffe number of bins to both plots
versicolor=attr[target==1,:]
num_bins=15
plt.hist(setosa[:,1],num_bins,alpha=.9)
plt.hist(versicolor[:,1],10,alpha=0.75)
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.show()
```

```python
#adding legends
versicolor=attr[target==1,:]
num_bins=15
plt.hist(setosa[:,1],num_bins,alpha=.9)
plt.hist(versicolor[:,1],10,alpha=0.75)
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.legend(['setosa','versicolor'])
plt.show()
```

```python
import pandas as pd
daily = pd.read_csv('day.csv')
daily.head()
```

```python
x=daily.loc[:,'day']
y=daily.loc[:,'cnt']
plt.plot(x,y,linewidth=2.5)
plt.xlabel('Day')
plt.ylabel('Check outs')
plt.show()
#insight: fluctuating is based on season in winter there are low checkouts
```

```python
#We will plot the daily count of bikes that were checked out by
#casual/non-registered users against the temperature. Color the points to be ‘#539cab’. Set
#the transparency to be 0.7. Be sure to include appropriate labels for x-axis and y-axis. What
#insight can you get?

x=daily.loc[:,'temp']
y=daily.loc[:,'casual']
plt.xlabel('temp')
plt.ylabel('Casual count')
plt.scatter(x,y,s=80,c='#539cab',marker='>',alpha=.7)
plt.show()











```

```python
x=daily.loc[:,'temp']
y=daily.loc[:,'casual']
plt.xlabel('temp')
plt.ylabel('Casual count')

plt.scatter(x,y,s=80,c=daily.loc[:,'workingday'],marker='>',alpha=.7)
cbar= plt.colorbar(ticks=[0,1])
cbar.set_ticklabels(['non-working day','Working'])
plt.show()


#insights much more linear relationships,if temp is higher 
```

```python

daily.loc[:,'season']==1
Winter=daily.loc[daily.loc[:,'season']==1,'temp']
Spring=daily.loc[daily.loc[:,'season']==2,'temp']
Summer=daily.loc[daily.loc[:,'season']==3,'temp']
Fall=daily.loc[daily.loc[:,'season']==4,'temp']
plt.hist(Winter,10)
plt.hist(Spring,10,alpha=.8)
plt.hist(Summer,10,alpha=.8)
plt.hist(Fall,10,alpha=.8)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.legend(['Winter','Spring','Summer','Fall'])
plt.show()

# in winter and summer ,temp is more concentrated is dipersion is less as temp does not vary much during these season
 #while thats not the case for fall or sprin
    # temo is highest for summer and lowest for summer
```
