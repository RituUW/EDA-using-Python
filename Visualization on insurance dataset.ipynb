```python
import pandas as pd
insurance = pd.read_csv('insurance.csv')
insurance.head()
```

```python
from matplotlib import pyplot as plt
smoker=insurance.loc[:,'smoker']
plt.scatter(insurance.loc[:,'bmi'],insurance.loc[:,'charges'],
            c=smoker,alpha=0.7)
plt.xlabel('Body Mass Index')
plt.ylabel('Medical Cost($)')
cbar= plt.colorbar(ticks=[0,1])
cbar.set_ticklabels(['non-smoker','smoker'])
plt.show()

# Obs1 :BMI does not have significant impact on medical cost for Non-Smokers however for Smokers, Medical cost 
#increases with increase in BMI
#obs2: Medical cost is higher for smoker as compared to non smokers irrespective of BMI
#Obs3:there are several NON SMOKERS with high medical cost which
#could be due to other attributes like age or gender etc .
```

```python
Smoker=insurance.loc[insurance.loc[:,'smoker']==1,'charges']
Non_Smoker=insurance.loc[insurance.loc[:,'smoker']==0,'charges']
num_bins=12
plt.hist(Smoker,num_bins)
plt.hist(Non_Smoker,num_bins,alpha=0.6)
plt.xlabel('Medical cost($)')
plt.ylabel('Number of Patients')
plt.legend(['Smoker','Non-Smoker'])
plt.title('Distribution of Medical cost(Smoker vs Non-smoker)')
plt.show()
#medical cost is lower for non smokers as compared to smokers.
#Majority of non smokers incur medical cost between 0 to 15000$
# medical cost with value 15000 or higher are mostly linked to smokers
# Also we see great variability in medical cost for smokers which could be due to other factors such as 
#BMI,age,gender,number of dependents or the region.
```

```python
Young=insurance.loc[insurance.loc[:,'age']<40,'charges']
Old=insurance.loc[insurance.loc[:,'age']>=40,'charges']
num_bins=15
plt.hist(Young,num_bins)
plt.hist(Old,num_bins,alpha=0.7)
plt.xlabel('Medical cost($)')
plt.ylabel('Number of Patients')
plt.title('Distribution of Medical Cost (Young vs Old)')
plt.legend(['Young Patients','Old Patients'])
plt.show()
# As we can see from the plot, majority of young people have lower medical cost as compared to old patients which make 
#sense as young people have lower health problems in comparison to old people.
#However,there is an exception to this: There are more young people with medical cost around 18000,32000 than old patients. 
   # which can be attributed to  other factors such as their smoking habit,BMI etc.
```

```python
Female=insurance.loc[insurance.loc[:,'sex']==1,'charges']
Male=insurance.loc[insurance.loc[:,'sex']==0,'charges']
num_bins=15
plt.hist(Female,num_bins)
plt.hist(Male,num_bins,alpha=0.7)
plt.xlabel('Medical cost ($)')
plt.ylabel('Number of patients')
plt.title('Distribution of Medical Cost (Female vs Male)')
plt.legend(['Female','Male'])
plt.show()

#Observation - Majority of population incur medical costs lower than 30K irrespective of sex. 
# However People with Medical costs higher than 30K are mostly Males.

#Insight - Males get more serious illness than Females.
```

```python
smoker=insurance.loc[:,'smoker']
plt.scatter(insurance.loc[:,'age'],insurance.loc[:,'charges'],
            c=smoker,alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Medical Cost ($)')
cbar= plt.colorbar(ticks=[0,1])
cbar.set_ticklabels(['non-smoker','smoker'])
plt.show()

#Observation - Medical cost relationship is linear with Age however people with Smoking habits has higher medical 
#cost irrespective of Age. For instance - There is huge difference in medical cost between smoker and non-smoker 
#paitents of Age 20. Difference in medical cost of non-smokers and smokers is comparable at each Age group.
```
